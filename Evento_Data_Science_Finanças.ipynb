```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
Base_Dados = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')

# Verificando
Base_Dados.head()
```

```python
# Dimensao
Base_Dados.shape
```

```python
# Tipo dos Dados
Base_Dados.info()
```

```python
# Tipo dos Dados
Base_Dados.describe()
```

```python
# Ajustando o índice
Dados = Base_Dados.set_index('Data')

# Verificando
Dados.head()
```

```python
# Estilo
plt.style.use('seaborn-dark')

# Tamanho
plt.figure( figsize=(16,5) )

# Plotando a ação espeficifico
plt.plot( Dados.index, Dados['Fechamento'], color='#004E9F', linewidth=2, alpha=0.7)

plt.legend( ['Fechamento'], ncol=4, fontsize=12, loc='upper center',
           bbox_to_anchor=(0.5, -0.1), shadow=True )

plt.xlabel('Período histórico')
plt.ylabel('Preço de fechamento das ações', fontsize=12)

plt.title('Fechamento ações da Magalu', 
          loc='center', fontsize=18, fontweight=0, color='gray');
```

```python
# Estilo
plt.style.use('seaborn-dark')

# Tamanho
plt.figure( figsize=(16,5) )

# Calculando a media móvel
Media_Movel = Dados['Fechamento'].rolling(5).mean()
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()

# Plotando a ação espeficifico
plt.plot( Dados.index, Dados['Fechamento'], color='#004E9F', linewidth=2, alpha=0.5)
plt.plot( Dados.index, Media_Movel, color='#00652E', linewidth=2, alpha=0.7 )
plt.plot( Dados.index, Media_Tendencia, color='#ED6E0C', linewidth=2, alpha=0.7 )

plt.legend( ['Fechamento', 'Média Móvel', 'Tendência'], ncol=4, fontsize=12, loc='upper center',
           bbox_to_anchor=(0.5, -0.1), shadow=True )

plt.xlabel('Período histórico')
plt.ylabel('Preço de fechamento das ações', fontsize=12)

plt.title('Fechamento ações da Magalu', 
          loc='center', fontsize=18, fontweight=0, color='gray');
```

```python
# Boxplot

# Estilo
plt.style.use('seaborn-dark')

# Tamanho
plt.figure( figsize=(16,5) )

# Plot do Boxplot
sns.boxplot( data=Dados, x='Fechamento' );
```

```python
# Boxplot

# Ajustando os Valores
Base_Dados['Mes'] = Base_Dados['Data'].dt.month

# Estilo
plt.style.use('seaborn-dark')

# Tamanho
plt.figure( figsize=(16,5) )

# Plot do Boxplot
sns.boxplot( data=Base_Dados, y='Fechamento', x='Mes' );
```

```python
# Analise da Tabela
Base_Dados.groupby(['Mes']).describe()['Fechamento']
```

```python
plt.plot( Base_Dados.groupby(['Mes']).describe()['Fechamento']['std'] )
```

```python
# Plot - Plotly

# Criando o gráfico
Grafico = go.Figure(
    data=[
          go.Candlestick(
              x= Dados.index,
              open = Dados['Abertura'],
              high= Dados['Maior'],
              low= Dados['Menor'],
              close= Dados['Fechamento'],
              increasing_line_color= 'green', 
              decreasing_line_color= 'red' ) ] )

# Retirando a legeda abaixo
Grafico.update_layout (xaxis_rangeslider_visible=False )

# Mostrando
Grafico.show()
```

```python
# Quanto ganhei

# Quantoas ações você comprou
Comprei = 1000

Quanto_Comprei = round( Base_Dados['Fechamento'].values[0] * Comprei, 0)
Quando_Acabou_Ano = round( Base_Dados['Fechamento'].values[-1] * Comprei, 0)
GanhoPerda = round( Quanto_Comprei - Quando_Acabou_Ano, 0 )

print( str(' *** Hora da verdade *** \n').upper() )
print('Você investiu: R$', Quanto_Comprei )
print('Fechou o ano com: R$', Quando_Acabou_Ano )

if GanhoPerda >= 0:
  print('Você perdeu: R$', GanhoPerda )

else:
  print('Você ganhou: R$', GanhoPerda )
```
